**Packet delay change according to test packet number**

```python
import numpy as np
import matplotlib.pyplot as plt

packet_Num = [1,10,20,100,300,500]
send_min = [52.5,50.2,56.4,57,62.6,63.9]
send_median = [52.5,51,57.1,57.6,63.5,64.9]
send_max = [52.5,53.3,58.4,61.2,67.4,68.8]
receive_min = [42.4, 47.1, 17, 18.8, 42.4, 28.3]
receive_median = [42.5,48.7,54.7,55.3,61.3,62.6]
receive_max = [42.6,49,55.1,55.9,61.8,63.4]

l_send_min = plt.plot(packet_Num, send_min,'r:', label = 'min delay (client ->server)')
l_send_median = plt.plot(packet_Num, send_median,'r--', label = 'median delay (client ->server)')
l_send_max = plt.plot(packet_Num, send_max,'r:', label = 'max delay (client ->server)')
l_receive_min = plt.plot(packet_Num, receive_min,'g:', label = 'min delay (server -> client)')
l_receive_median = plt.plot(packet_Num, receive_median,'g--', label = 'median delay (server -> client)')
l_receive_max = plt.plot(packet_Num, receive_max,'g:', label = 'max delay (server -> client)')
plt.plot(packet_Num, send_median, 'ro--', packet_Num, receive_median, "bo--")
plt.title("Packet delay change according to test packet number")
plt.xlabel('Packet number')
plt.ylabel('delay(ms)')
plt.legend()
plt.show()
```
